# importing data

```python
import pandas as pd  # Importing pandas for data manipulation
import numpy as np  # Importing numpy for numerical operations

# Loading a dataset from a pickle file
df = pd.read_pickle("F:\programming\machine learning\quantum entangelment data")  # Reads a .pkl (pickle) file from the given path

# Printing the dataframe to check its contents
print(df)
```

# dividing data to test and train

```python
# Selecting the first 7000 rows as the training dataset
train = df.head(7000)  

# Selecting the last 3000 rows as the test dataset
test = df.tail(3000)  

# Printing the training dataset
print(train)
```

# creating test and train

```python
# Extracting the first 7000 labels (target values) for training
trainy = df['is_entangled'].to_numpy()[:7000]  

# Initializing an empty list for training features
trainx = []

# Looping through the first 7000 rows to extract the first 15 features
for i in range(7000):
    trainx.append(df.iloc[i].to_numpy()[:15])  

# Printing the training features
print(trainx)

# Extracting the remaining labels (target values) for testing
testy = df['is_entangled'].to_numpy()[7000:]  

# Initializing an empty list for test features
testx = []

# Looping from index 7000 to 9999 (assuming df has at least 10,000 rows)
for i in range(7000, 10000):
    testx.append(df.iloc[i].to_numpy()[:15])  
```

# LogisticRegression

```python
from sklearn import linear_model  # Importing linear models from sklearn
from sklearn.metrics import accuracy_score  # Importing accuracy metric

# Initializing Logistic Regression model
reg = linear_model.LogisticRegression()

# Training the model on training data
reg.fit(trainx, trainy)  

# Making predictions on test data
predictions = reg.predict(testx)  

# Calculating and printing the accuracy of the model
print("Accuracy:", accuracy_score(testy, predictions))  
```

# RidgeClassifier

```python
from sklearn.linear_model import RidgeClassifier  # Import RidgeClassifier for classification
from sklearn.model_selection import train_test_split  # For data splitting (if needed)
from sklearn.metrics import accuracy_score  # For evaluating model performance
import numpy as np  # Ensure data is in NumPy format

# Initialize RidgeClassifier with alpha (regularization strength)
ridge_clf = RidgeClassifier(alpha=12)  

# Train the classifier on training data
ridge_clf.fit(trainx, trainy)

# Make predictions on the test set
y_pred = ridge_clf.predict(testx)

# Evaluate accuracy
print("Accuracy:", accuracy_score(testy, y_pred))
```
