```python
from PIL import Image                         # Import PIL for image processing
import numpy as np                            # Import NumPy for numerical operations
import matplotlib.pyplot as plt               # Import matplotlib for plotting
from sklearn.decomposition import PCA         # Import PCA  for dimensionality reduction
# image data path : https://drive.google.com/drive/folders/1GiM_pE5pkHU-OJIoyV2G9Mq4fX_0yCej

# Load the image from the specified path
image = Image.open("F:\programming\ex2\Google-trex.png") 

image = image.convert("L")                    # Convert the image to grayscale ("L" mode)

# Convert image to NumPy array ( a matrix of pixel values)
image_matrix = np.array(image)

# Print the shape of the image (rows x columns)
print("Image shape:", image_matrix.shape)

# Print the actual matrix (grayscale pixel values)
print("Top-left 5x5 pixels:\n", image_matrix[:5, :5])

# Display the grayscale image
plt.imshow(image_matrix, cmap='gray')         # Show image with grayscale colormap
plt.title("Original Image")                   # Set the title of the plot
plt.axis("off")                               # Hide axis for better visualization
plt.show()                                    # Display the plot
```

```python
num_components = 5                            # Number of principal components to keep
pca = PCA(n_components=num_components)        # Create a PCA object with specified components

compressed = pca.fit_transform(image_matrix)  # Apply PCA to reduce dimensionality 
reconstructed = pca.inverse_transform(compressed)  # Reconstruct the image from the compressed version

# Show reconstructed image
plt.imshow(reconstructed, cmap='gray')        # Display the reconstructed image in grayscale
plt.title(f"Reconstructed Image with {num_components} Components")  # Title showing number of components
plt.axis("off")                               # Hide axes
plt.show()                                    # Show the plot
```
